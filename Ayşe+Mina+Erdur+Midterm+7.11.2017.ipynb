```python
import numpy as np


L = 200
list= []
for x in range(0,L):
  list.append(np.random.randint(-9,9))

MA = []

for x in range(1, len(list)+1):
  average = 0
  for i in range(x, 0, -1):
      average = average + list[i-1]

MA.append(round(average/len(list),2))

print(list)
print(MA)

import matplotlib.pyplot as plt
plt.plot(list)
```

```python
#European Call

import numpy as np
import random

tbegin = 0
tend = 5
deltat = .0005
t = np.arange(tbegin, tend, deltat)

E=1.0

sigma = 0.3
S0 = 1

sqrtdt = np.sqrt(deltat)

N=np.zeros(1000) 
for i in range(0,N.size):
    y = S0
    for j in range(1, t.size):
        y = y * np.exp(random.choice([-1,1]) * sigma * sqrtdt)
    N[i]=y

N=np.maximum(0,N-E)
print(N.mean())
```

```python
#European Put

import numpy as np
import random

tbegin = 0
tend = 1
deltat = .0001
t = np.arange(tbegin, tend, deltat)

E=1.0

sigma = 0.3
S0 = 1

sqrtdt = np.sqrt(deltat)

N=np.zeros(1000) 
for i in range(0,N.size):
    y = S0
    for j in range(1, t.size):
        y = y * np.exp(random.choice([-1,1]) * sigma * sqrtdt)
    N[i]=y

N=np.maximum(0,E-N)
print(N.mean())
```
